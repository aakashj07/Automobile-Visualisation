```python
#To load dataset automobile_data
car=pd.read_csv("Automobile_data.csv")  #Here car user defined dataframe
car.head() #To print first 5 record
```

```python
sns.countplot(data=car,x="fuel-type")
#Syntax sns.countplot(from where data came,column input)
f=car["fuel-type"].value_counts()
plt.yticks(f)
plt.show()
```

```python
car["body-style"].value_counts()
```

```python
plt.figure(figsize=(7,12))
sns.countplot(data=car,x="fuel-type",hue="body-style")
#hue inbuilt parameter, it is used to subgroup column
f=car["body-style"].value_counts()
plt.yticks(f)
plt.show()
```

```python
plt.figure(figsize=(7,12))
sns.countplot(data=car,x="fuel-type",hue="body-style")
#hue inbuilt parameter, it is used to subgroup column
f=car["body-style"].value_counts()
plt.title("fuel-type & body-style")
plt.yticks(f)
plt.show()
```

```python
plt.figure(figsize=(7,8))
sns.countplot(data=car,x="body-style")
#hue inbuilt parameter, it is used to subgroup column
f=car["body-style"].value_counts()
plt.title("fuel-type & body-style")
plt.yticks(f)
plt.show()
```

```python
#pie graph
car["body-style"].value_counts().plot(kind="pie",autopct="%.1f%%")
#value*100/360   -->convertible==>6/360*100
#%d means integer 
#%f means float means show after decimal 6 digits
#%.1f means float value show but after decimal only 1 digit show
#dataframe["categorical column"].value_count().plot(kind="pie")
plt.show()
```

```python
#pie graph
car["body-style"].value_counts().plot(kind="pie",autopct="%d%%")
#d=int
#value*100/360   -->convertible==>6/360*100
#dataframe["categorical column"].value_count().plot(kind="pie")
plt.show()
```

```python
#pie graph
car["body-style"].value_counts().plot(kind="pie",autopct="%.1f%%",explode=(0,0,0,0.5,0.7))
#value*100/360   -->convertible==>6/360*100
#%d means integer 
#%f means float means show after decimal 6 digits
#%.1f means float value show but after decimal only 1 digit show
#dataframe["categorical column"].value_count().plot(kind="pie")
plt.show()
```
